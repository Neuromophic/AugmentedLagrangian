```python
import torch
import numpy as np
```

```python
hes_p = torch.load('./penalty.hessian')
hes_a = torch.load('./al.hessian')
```

```python
import matplotlib.pyplot as plt

hes = torch.zeros([13,4,2])
for ds in range(13):
    for i in range(4):
        plt.figure()
        plt.plot(hes_p[ds,i,:], color='blue')
        plt.plot(hes_a[ds,i,:], color='orange')
        hes[ds,i,0] = hes_p[ds,i,:].max()
        print(hes_p[ds,i,:].max())
        hes[ds,i,1] = hes_a[ds,i,:].max()
        plt.show()
```

```python
hes = hes.numpy()
```

```python
for ds in range(13):
    filename = f"hessian_slice_{ds:02d}.txt"
    np.savetxt(filename, hes[ds], fmt="%.3e", delimiter='\t')

    # Print out the saved slice for verification (optional)
    print(f"Saved hes[{ds}] to '{filename}'")
```
