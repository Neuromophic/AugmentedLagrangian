```python
import torch
import pickle
```

```python
datapath = './dataset/Dataset_pendigits.p'
```

```python
with open(datapath, 'rb') as f:
    data = pickle.load(f)
```

```python
data
```

```python
X_train = data['X_train']
y_train = data['y_train']
X_valid = data['X_valid']
y_valid = data['y_valid']
X_test = data['X_test']
y_test = data['y_test']

X_all = data['X_all']
y_all = data['y_all']
n_feature = data['n_feature']
n_class = data['n_class']
name = data['name']
```

```python
X_train.shape
```

```python
def SS(X, y, n=3):
    subset_size = X.shape[0] // n
    indices = torch.randperm(X.shape[0])    
    subset_indices = indices[:subset_size]
    X_subset = X[subset_indices]
    y_subset = y[subset_indices]
    return X_subset, y_subset
```

```python
X_train, y_train = SS(X_train,y_train)
X_train.shape, y_train.shape
```

```python
X_valid, y_valid = SS(X_valid,y_valid)
X_valid.shape, y_valid.shape
```

```python
X_test, y_test = SS(X_test,y_test)
X_test.shape, y_test.shape
```

```python
data['X_train'] = X_train
data['y_train'] = y_train
data['X_valid'] = X_valid
data['y_valid'] = y_valid
data['X_test']  = X_test
data['y_test']  = y_test
```

```python
save_path = './dataset/Dataset_pendigits_small.p'

with open(save_path, 'wb') as f:
    pickle.dump(data, f)
```
